```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd  # <--- Esto es lo que genera el Excel
import time

# 1. Configuraci√≥n e Inicio
print("ü§ñ Iniciando el robot...")
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# 2. Navegar a Google
driver.get("https://www.google.com")
time.sleep(3)

# 3. Buscar "Lenovo Legion"
print("üîç Buscando en Google...")
caja_busqueda = driver.find_element(By.NAME, "q")
caja_busqueda.send_keys("Lenovo Legion precio peru") 
caja_busqueda.send_keys(Keys.RETURN)
time.sleep(4) # Esperamos a que carguen los resultados

# 4. EXTRAER DATOS (Web Scraping)
print("‚õèÔ∏è Extrayendo t√≠tulos de los resultados...")
# Buscamos todos los elementos h3 (t√≠tulos de Google)
resultados = driver.find_elements(By.TAG_NAME, "h3")

datos_extraidos = []

for resultado in resultados:
    titulo = resultado.text
    # Solo guardamos si tiene texto (a veces hay elementos vac√≠os ocultos)
    if len(titulo) > 0:
        datos_extraidos.append(titulo)
        print(f"   -> Encontrado: {titulo}")

# 5. GENERAR EXCEL
print("üíæ Guardando archivo Excel...")
df = pd.DataFrame(datos_extraidos, columns=["T√≠tulos Encontrados"])
df.to_excel("reporte_precios.xlsx", index=False)

print("‚úÖ ¬°Misi√≥n Cumplida! Revisa el archivo 'reporte_precios.xlsx'.")
driver.quit()
```
